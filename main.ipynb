# Redução de dimensionalidade

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

## Funções de redução de dimensionalidade

```python
# Carregar imagem colorida
def load_image(path):
    return Image.open(path)

# Converter para tons de cinza
def rgb_to_grayscale(img):
    img_array = np.array(img)
    r, g, b = img_array[:, :, 0], img_array[:, :, 1], img_array[:, :, 2]
    gray_array = 0.299 * r + 0.587 * g + 0.114 * b
    return Image.fromarray(gray_array.astype(np.uint8))

# Binarizar imagem
def grayscale_to_binary(img, threshold=128):
    img_array = np.array(img)
    binary_array = np.where(img_array > threshold, 255, 0)
    return Image.fromarray(binary_array.astype(np.uint8))
```

```python
# Carregar imagem colorida
color_image = load_image("img/lena_colorida.jpeg")
```

```python
# mostrar imagem colorida via matplotlib
plt.imshow(color_image)
plt.title("Imagem Colorida")
plt.show()
```

```python
# Converter para tons de cinza
gray_image = rgb_to_grayscale(color_image)
gray_image.save("img/lena_grayscale.jpg")
```

```python
# mostrar imagem em tons de cinza via matplotlib
plt.imshow(gray_image, cmap="gray")
plt.title("Imagem em Tons de Cinza")
plt.show()
```

```python
# Converter para binária
binary_image = grayscale_to_binary(gray_image)
binary_image.save("img/lena_binary.jpg")
```

```python
# mostrar imagem binária via matplotlib
plt.imshow(binary_image, cmap="gray")
plt.title("Imagem Binária")
plt.show()
```
